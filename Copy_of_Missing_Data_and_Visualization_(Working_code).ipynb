# Handling missing data

### The first method to handle missing values could be to simply delete the rows with any NaN values. 

```python
updated_df1 = hospital_df.dropna()
print(updated_df1.isnull().sum())
```

```python
print(updated_df1.iloc[0:20,:])
```

```python
updated_df1
```

```python
# To reset index of updated dataframe
updated_df2 = hospital_df.dropna()
updated_df2.reset_index(inplace=True)
print(updated_df2.isnull().sum())
```

```python
print(updated_df2.iloc[0:20,:])
```

```python
updated_df2
```

```python
updated_df2.shape
```

### The second method is to impute the missing values with the mean of that particular column.

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
updated_df3 = imp.fit_transform(hospital_df)
updated_df3 = pd.DataFrame(updated_df3, columns=hospital_df.columns)
```

```python
print(updated_df3.isnull().sum())
```

```python
updated_df3
```

```python
print(updated_df3.iloc[[15,22,37],:])
```

### The median, mode or a constant can also be used to impute the missing values of a column. This can be done by changing the 'strategy' parameter of the SimpleImputer function. 

```python
#Replace all missing values in the dataset with 0.

imp = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0)
updated_df4 = imp.fit_transform(hospital_df)
updated_df4 = pd.DataFrame(updated_df4, columns=hospital_df.columns)
print(updated_df3.isnull().sum(), '\n')
print(updated_df4.iloc[[15,22,37],:])
```

```python
# Imputing a single column AGE with its mean

imp = SimpleImputer(missing_values=np.nan,strategy='mean')
X = np.array(hospital_df['AGE']).reshape(-1,1)
X_updated = imp.fit_transform(X)
print(X_updated)
X_updated = X_updated.reshape(1,-1)[0]
X_updated==np.nan
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer 
```

```python
imp2 = IterativeImputer()
updated_df8 = imp2.fit_transform(hospital_df)
updated_df8 = pd.DataFrame(updated_df8, columns=hospital_df.columns)
print(updated_df8.isnull().sum(), '\n')
print(updated_df8.iloc[[15,22,37],:])
```

### The third method is an imputation using MICE (Multiple Imputation with Chained Equations) 
Ref: https://www.statsmodels.org/dev/imputation.html

```python
import statsmodels.imputation.mice as sm
```

```python
imp = sm.MICEData(hospital_df)
imp.update_all()
updated_df5 = imp.data
updated_df5.isnull().sum()
```

```python
print(updated_df5.iloc[[15,22,37],:])
```

### The fourth method is an imputation using k-Nearest Neighbors 

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=1)
updated_df6 = imputer.fit_transform(hospital_df)
updated_df6 = pd.DataFrame(data=updated_df6, columns = hospital_df.columns )
updated_df5.isnull().sum()
```

```python
print(updated_df6.iloc[[15,22,37],:])
```
